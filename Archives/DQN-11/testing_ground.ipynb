```python
import numpy as np
from itertools import combinations_with_replacement
```

```python
x = [-0.01, 0, 0.01]
n_asset = 3
```

```python
action_possible = np.array(np.meshgrid(*([x] * n_asset))).T.reshape(-1, n_asset)
action_possible = action_possible[action_possible.sum(axis=1) == 0, :]
```

```python
action_possible
```

```python
n = 10
state_possible = np.array(list(combinations_with_replacement(x, n)))
state_possible
```
